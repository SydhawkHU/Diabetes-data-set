```python
import kagglehub
import pandas as pd
import numpy as np
import plotly.express as px
import os

# Download
path = kagglehub.dataset_download("marshalpatel3558/diabetes-prediction-dataset")

# Find CSV
for f in os.listdir(path):
    if f.endswith(".csv"):
        csv_path = os.path.join(path, f)
        break

df = pd.read_csv(csv_path)
```

```python
fig = px.scatter(
    tips,
    x="Age",
    y="Blood_Pressure_Systolic",
    color="Sex",   # seperated by gender
    color_discrete_map={
        "Male": "blue",
        "Female": "pink"
    },
    width=700,
    height=450,
    labels={
        "Age": "Age (years)",
        "Blood_Pressure_Systolic": "Systolic Blood Pressure",
        "Sex": "Gender"
    },
    title="Age vs Systolic Blood Pressure (Colored by Gender)"
)

fig.update_traces(marker=dict(size=10))
fig.show()
```

```python
# Filter systolic BP between 100 and 140
filtered = tips[
    (tips["Blood_Pressure_Systolic"] >= 100) &
    (tips["Blood_Pressure_Systolic"] <= 140)
]

fig = px.scatter(
    filtered,
    x="Age",
    y="Blood_Pressure_Systolic",
    color="Sex",
    color_discrete_map={
        "Male": "blue",
        "Female": "pink"
    },
    width=700,
    height=450,
    labels={
        "Age": "Age (years)",
        "Blood_Pressure_Systolic": "Systolic Blood Pressure"
    },
    title="Age vs Systolic Blood Pressure (100140 Range, Colored by Gender)"
)

fig.update_traces(marker=dict(size=10))
fig.show()
```
